**Histogram Analizi**

```python
import wave
import numpy as np
import matplotlib.pyplot as plt


ses_dosyasi = "/kaggle/input/audio-cats-and-dogs/cats_dogs/test/cats/cat_110.wav" 
ses = wave.open(ses_dosyasi, 'rb')

#dosyayı oku
params = ses.getparams()
n_frames = params.nframes
audio_frames = np.frombuffer(ses.readframes(n_frames), dtype=np.int16)

# histogram grafigi
plt.hist(audio_frames, bins=50, color='blue', alpha=0.7)
plt.title("Ses Dosyasındaki Histogram")
plt.xlabel("Amplitude")
plt.ylabel("Frekans")
plt.show()
```

**Frekans Domaini Analizi**

```python
frekans_bilesenleri = np.fft.fft(audio_frames)
frekanslar = np.fft.fftfreq(len(audio_frames))


plt.plot(frekanslar[:len(frekanslar)//2], np.abs(frekans_bilesenleri)[:len(frekans_bilesenleri)//2])
plt.title("Ses Dosyasının Frekans Spektrumu")
plt.xlabel("Frekans (Hz)")
plt.ylabel("Büyüklük")
plt.show()
```

**Zaman Domaini Analizi**

```python
plt.plot(audio_frames[:1000]) 
plt.title("Zaman Domainindeki Ses Dalga Formu")
plt.xlabel("Örnek Numarası")
plt.ylabel("Genlik")
plt.show()
```
