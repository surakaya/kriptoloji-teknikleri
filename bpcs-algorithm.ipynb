```python
import numpy as np
from PIL import Image
import cv2
```

```python
gorsel_yolu = "/kaggle/input/mnistasjpg/testSample/testSample/img_101.jpg" 
img = cv2.imread(gorsel_yolu, cv2.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Görsel yüklenemedi. Dosya yolunu kontrol edin: {gorsel_yolu}")
else:
    print("Görsel başarıyla yüklendi.")
```

```python
def bpcs_gorsel_analizi(gorsel_yolu, mesaj, cikti_gorsel):
   
   
    mesaj += "####"
    mesaj_bitleri = ''.join([format(ord(karakter), '08b') for karakter in mesaj])
    
   
    bit_plane = [((img >> i) & 1) for i in range(8)]
    
   
    karmaşik_bolgeler = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
          
            complexity = sum([bit_plane[k][i, j] for k in range(8)]) 
            if complexity > 4: 
                karmaşik_bolgeler.append((i, j))
    
    
    bit_index = 0
    for (i, j) in karmaşik_bolgeler:
        if bit_index < len(mesaj_bitleri):
           
            current_bit = int(mesaj_bitleri[bit_index])
            img[i, j] = (img[i, j] & ~1) | current_bit
            bit_index += 1
        else:
            break

  
    cv2.imwrite(cikti_gorsel, img)
    print(" Mesaj başarıyla BPCS yöntemiyle gömüldü:", cikti_gorsel)


def mesaj_coz_bpcs(gorsel_yolu):
    img = cv2.imread(gorsel_yolu, cv2.IMREAD_GRAYSCALE)
    
    if img is None:
        print(f"Error: Görsel yüklenemedi. Dosya yolunu kontrol edin: {gorsel_yolu}")
        return
    
    
    bit_plane = [((img >> i) & 1) for i in range(8)]
    
    bitler = ""
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
           
            complexity = sum([bit_plane[k][i, j] for k in range(8)])
            if complexity > 4: 
                bitler += str(bit_plane[7][i, j])  
    
    
    mesaj = ""
    for i in range(0, len(bitler), 8):
        byte = bitler[i:i+8]
        if len(byte) == 8:
            mesaj += chr(int(byte, 2))
            if "####" in mesaj:
                print("Çözülen Mesaj:")
                print(mesaj.split("####")[0])
                return

    print(" Mesaj bulunamadı.")


gorsel_yolu = "ornek_gorsel.jpg"
mesaj = "Bu görselin içine BPCS yöntemiyle gizlenmiş bir mesaj bulunmaktadır."
bpcs_gorsel_analizi(gorsel_yolu, mesaj, "gizli_bpcs.jpg")
mesaj_coz_bpcs("gizli_bpcs.jpg")
```
